```python
from preprocessed_mnist import load_dataset
X_train, y_train, X_val, y_val, X_test, y_test = load_dataset()
print(X_train.shape, y_train.shape)
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(X_train[0], cmap="Greys");
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

X_train = X_train.reshape(X_train.shape[0], -1).astype('float32')
X_val = X_val.reshape(X_val.shape[0], -1).astype('float32')
X_test = X_test.reshape(X_test.shape[0], -1).astype('float32')

m, n = X_train.shape
n_classes = len(set(y_train))

sess = tf.Session()
y_train = sess.run(tf.one_hot(y_train, n_classes))
y_test = sess.run(tf.one_hot(y_test, n_classes))
y_val = sess.run(tf.one_hot(y_val, n_classes))
sess.close()

print(X_train.shape)
print(y_train.shape)
print(y_test.shape)
print(y_val.shape)
```

```python
def forward_prop(X, w1, b1, w2, b2):
    z1 = tf.add(X @ w1, b1)
    a1 = tf.nn.sigmoid(z1)
    z2 = tf.add(a1 @ w2, b2)
#     z2 = tf.nn.sigmoid(z2)
    
    return z2


def loss(z2, Y, w1, w2, lamda=0):
    j = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=z2, labels=Y))  + lamda * (tf.nn.l2_loss(w1) + tf.nn.l2_loss(w2))
    return j

```

```python
alpha = 0.01
lamda = 0
layer_size = 70
epochs = 500

tf.set_random_seed(777)
x = tf.placeholder(tf.float32, [None, n])
y = tf.placeholder(tf.float32, [None, n_classes])
w1 = tf.Variable(initial_value=tf.random_normal([n, layer_size]), dtype=tf.float32)
w2 = tf.Variable(initial_value=tf.random_normal([layer_size, n_classes]), dtype=tf.float32)
b1 = tf.Variable(initial_value=tf.random_normal([1, layer_size]), dtype=tf.float32)
b2 = tf.Variable(initial_value=tf.random_normal([1, n_classes]), dtype=tf.float32)

z2 = forward_prop(x, w1, b1, w2, b2)
cost = loss(z2, y, w1, w2, lamda=lamda)
accuracy = 100 * tf.reduce_mean(tf.cast(tf.equal(tf.argmax(z2, axis=1), tf.argmax(y, axis=1)), tf.float32))

optimizer = tf.train.AdamOptimizer(learning_rate = alpha).minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

j_history = []

for epoch in range(epochs):
    _, j, acc = sess.run([optimizer, cost, accuracy], feed_dict={x: X_train, y: y_train})
    acc_test = sess.run(accuracy, feed_dict={x: X_test, y: y_test})
    j_history.append(j)
    print("Epoch: %d\tCost: %0.3f\tTrain Accuracy: %0.2f\tTest Accuracy: %0.2f" % (epoch + 1, j, acc, acc_test))

plt.plot(range(1, epochs + 1), j_history)
plt.xlabel("No. of Iterations")
plt.ylabel("Cost")
plt.show()

parameters = sess.run([w1, b1, w2, b2])
w1, b1, w2, b2 = parameters[0], parameters[1], parameters[2], parameters[3]

print("Training accuracy: %0.2f\tTest set accuracy: %0.2f\tValidation set accuracy: %0.2f"
      % (sess.run(accuracy, feed_dict={x: X_train, y: y_train}), 
        sess.run(accuracy, feed_dict={x: X_test, y: y_test}), 
        sess.run(accuracy, feed_dict={x: X_val, y: y_val})))
```

```python
acc_gross = (sess.run(accuracy, feed_dict={x: X_train, y: y_train}) 
       + sess.run(accuracy, feed_dict={x: X_test, y: y_test}) 
       + sess.run(accuracy, feed_dict={x: X_val, y: y_val})) / 3
print("Gross Accuracy: %0.2f %%" % (acc_gross))
```
